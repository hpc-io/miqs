```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt


data = ((0.003062535,0.893186098,0.157687037), 
        (0.005696092,0.715854759,0.383640615), 
        (0.000689041,1.287960457,1.090298545))

dim = len(data[0])
w = 0.75
dimw = w / dim

fig, ax = plt.subplots()
x = np.arange(len(data))
for i in range(len(data[0])):
    y = [d[i] for d in data]
    b = ax.bar(x + i * dimw, y, dimw, bottom=0.001)

# ax.set_xticks()
ax.set_yscale('log')

ax.set_xlabel('Dataset')
ax.set_ylabel('CV of Keyword Distribution')

plt.show()
```
