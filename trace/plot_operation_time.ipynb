```python
internal_count_num_unindexed_matched = np.loadtxt('query_internal_count_1m_num_unindexed_matched.txt', delimiter = ',', dtype=float);

internal_count_num_unindexed_unmatched = np.loadtxt('query_internal_count_1m_num_unindexed_unmatched.txt', delimiter = ',', dtype=float);

external_count_num_unindexed_matched = np.loadtxt('query_external_count_1m_num_unindexed_matched.txt', delimiter = ',', dtype=float);

external_count_num_unindexed_unmatched = np.loadtxt('query_external_count_1m_num_unindexed_unmatched.txt', delimiter = ',', dtype=float);


internal_count_num_indexed_matched = np.loadtxt('query_internal_count_1m_num_indexed_matched.txt', delimiter = ',', dtype=float);

internal_count_num_indexed_unmatched = np.loadtxt('query_internal_count_1m_num_indexed_unmatched.txt', delimiter = ',', dtype=float);

external_count_num_indexed_matched = np.loadtxt('query_external_count_1m_num_indexed_matched.txt', delimiter = ',', dtype=float);

external_count_num_indexed_unmatched = np.loadtxt('query_external_count_1m_num_indexed_unmatched.txt', delimiter = ',', dtype=float);



xlabels1 = ['Matched', 'Unmatched']

xlabels2 = ['Matched', 'Unmatched']

data1 = np.column_stack((
    internal_count_num_unindexed_matched,
    internal_count_num_unindexed_unmatched
                       ))

data2 = np.column_stack((
    internal_count_num_indexed_matched,
    internal_count_num_indexed_unmatched
                       ))

data3 = np.column_stack((
    external_count_num_unindexed_matched,
    external_count_num_unindexed_unmatched
                       ))

data4 = np.column_stack((
    external_count_num_indexed_matched,
    external_count_num_indexed_unmatched
                       ))

fig, ax = plt.subplots(2,2)

ax[0][0].boxplot(data1,  showfliers=False)

ax[0][0].set(xlabel='Condition', ylabel='Time (s)', 
       title='Internal Count Query Performance (No Index)')
ax[0][0].set_xticklabels(xlabels1, rotation=0)

ax[1][0].boxplot(data2,  showfliers=False)

ax[1][0].set(xlabel='Condition', ylabel='Time (s)', 
       title='Internal Count Query Performance (With Index)')
ax[1][0].set_xticklabels(xlabels2, rotation=0)

ax[0][1].boxplot(data3,  showfliers=False)

ax[0][1].set(xlabel='Condition', ylabel='Time (s)', 
       title='External Count Query Performance (No Index)')
ax[0][1].set_xticklabels(xlabels2, rotation=0)

ax[1][1].boxplot(data4,  showfliers=False)

ax[1][1].set(xlabel='Condition', ylabel='Time (s)', 
       title='External Count Query Performance (With Index)')
ax[1][1].set_xticklabels(xlabels2, rotation=0)


plt.subplots_adjust(hspace=0.7, wspace=0.7)
```

```python
index_num_arr = np.arange(0, 6)

data_list=[]
filename_template="insert_1k_with_%d_index.txt"

for it in index_num_arr:
    filename=filename_template % it
    data_list.append(np.loadtxt(filename, delimiter=',', dtype=float))
    
data = np.column_stack(tuple(data_list))

fig, ax = plt.subplots(1,1)

ax.boxplot(data, showfliers=False)

ax.set(xlabel='Number of Indexed Fields', ylabel='Time (ms)', 
       title='Insert Performance')
ax.set_xticklabels(index_num_arr, rotation=0)
```
